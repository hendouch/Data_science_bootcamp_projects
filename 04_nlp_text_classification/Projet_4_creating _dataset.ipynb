```python
import os
import requests
from bs4 import BeautifulSoup
a ={
#'Ed_Sheeran':'https://www.lyrics.com/artist/Ed-Sheeran/2342870',
#'Eminem':'https://www.lyrics.com/artist/Eminem/347307',
'SB_nineteen':'https://www.lyrics.com/artist/SB19/2137849178',
#'Billie_Eilish':'https://www.lyrics.com/artist/Billie-Eilish/3177510',
#'Ariana_Grande':'https://www.lyrics.com/artist/Ariana-Grande/1223462',
'BTS':'https://www.lyrics.com/artist/BTS/2838463',
'Cardi-B':'https://www.lyrics.com/artist/Cardi-B/3195462',
   }
```

```python
artist=[]
songs=[]
link=[]
for url in a:
    art_html = requests.get(a[url]).text
    os.makedirs(url, exist_ok=True)
    file = open(file= "{}/{}.txt".format(url,url),mode="w")
    #file = open(file= f"{url}/{url}.txt",mode="w")
    file.write(art_html)
    soup=BeautifulSoup(art_html)
    #artist.append(soup.head.title.text)
    #artist.append(url)
    for links in soup.find_all(class_='tal qx'):
        for subtag in links.find_all('a'):
            link.append("https://www.lyrics.com"+subtag.get('href'))
            songs.append(subtag.text)
            artist.append(url)
            
print(len(songs))
print(len(link))
print(len(artist))
```

```python
import time
lyrics=[]
for url in link:
    song_html = requests.get(url)
    #time.sleep(3)
    song_soup = BeautifulSoup(song_html.text)
    if (song_html.status_code==200):
        try:
            lyrics.append(song_soup.find('pre').text)
            #file = open(file= "{}/{}.txt".format(y,j),mode="w")
            #file.write(song_soup.find('pre').text)
            #file.close()
        except:
            lyrics.append(song_soup.find('pre'))
            #file = open(file= "{}/{}.txt".format(y,j),mode="w")
            #file.write(song_soup.find('pre'))
            #file.close()
    else:
        break
#for i in range(len(link)):
    #print(lyrics[i])
len(lyrics)
```

```python
import pandas as pd
lyrics_df = pd.DataFrame({'Song_title':songs,'Links':link, 'Lyrics':lyrics, 'Artist_name':artist})
lyrics_df
```
