```python
df = pd.merge(left=fert, right=pop, on=['country','year'])
df
```

```python
df = pd.merge(left=life, right=df, on=['country','year'])
df
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_subset = df.loc[df['year'] == 2000]
sns.scatterplot(x='Life expectancy', y='fertility_rate', data=df_subset, alpha=0.6)
```

```python
2015-1960
```

```python
xmin= df["Life expectancy"].min()
xmax= df["Life expectancy"].max()
ymin= df["fertility_rate"].min()
ymax= df["fertility_rate"].max()
```

```python
for each_year in range(1960, 2015):
    df_subset = df.loc[df['year'] ==  each_year]
    sns.scatterplot(x='Life expectancy', y='fertility_rate', data=df_subset, alpha=0.6)
    plt.axis((xmin, xmax, ymin, ymax))
    plt.savefig('lifeexp_{}.png'.format(each_year), dpi=150)
plt.close()
```

```python
import imageio

images = []

for i in range(1960, 2015):
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=20)
```
