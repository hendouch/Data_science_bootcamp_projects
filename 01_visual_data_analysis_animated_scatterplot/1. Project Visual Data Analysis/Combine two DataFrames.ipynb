```python
import pandas as pd
import seaborn as sn
```

```python
df = pd.read_excel("gapminder_lifeexpectancy.xlsx")
df
```

```python
df1 = df[['Life expectancy', 2016]]
df1
```

```python
df1.rename(columns={'Life expectancy':'country'}, inplace=True)
df1
```

```python
df2 = pd.read_csv("continents.csv", sep=';')
df2
```

```python
final_df = pd.merge(left=df1, right=df2, on='country')
final_df
```

```python
final_df.rename(columns={2016:'Life expectancy in 2016'}, inplace=True)
final_df
```

```python
final_df1 = final_df.dropna()
final_df1
```

```python
maxi = final_df1.groupby('continent').max()
maxi
```

```python
maxi.boxplot()
```

```python
sn.boxplot(data = maxi)
```

```python
mini = final_df1.groupby('continent').min()
mini
```

```python
sn.boxplot(data = mini)
```
