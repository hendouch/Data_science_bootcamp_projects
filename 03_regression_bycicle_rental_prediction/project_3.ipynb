```python
df = df[df['weather'] != 4]
```

```python
df = df[df['workingday'] != 1]
```

```python
from sklearn.linear_model import LinearRegression
m = LinearRegression()
```

```python
df
```

```python
X = df.drop("count", axis=1)
y = df['count']
X.shape, y.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
from sklearn.preprocessing import PolynomialFeatures
from fast_ml.feature_engineering import FeatureEngineering_DateTime
transformer = ColumnTransformer(
    transformers=[
        ('trans',PolynomialFeatures(degree=3, include_bias=False),['temp','humidity', 'windspeed']),
        #('trans1',MinMaxScaler(),['month','day']),
        ('trans2',OneHotEncoder(drop='first'),['year','month','day']),
        ],
        remainder = 'drop')
```

```python
from sklearn.model_selection import train_test_split,cross_val_score,KFold,StratifiedKFold
Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,test_size=0.2)
Xtrain.shape, Xtest.shape
```

```python
Xtrain_preprocess = transformer.fit_transform(Xtrain)
```

```python
m.fit(Xtrain_preprocess,ytrain)
```

```python
print(f'The training score is {m.score(Xtrain_preprocess,ytrain)}')
```

```python
Xtest_preprocess = transformer.fit_transform(Xtest)
```

```python
scores = cross_val_score(m,Xtrain_preprocess,ytrain,cv=KFold(5,shuffle=True),scoring='r2') # Kfold used for regression problems
# StratifiedKfold used for classification problems
scores 
```

```python
m.fit(Xtest_preprocess,ytest)
```

```python
ytest_predict=m.predict(Xtest_preprocess)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(ytest, ytest_predict)
mse
```

```python
scores.mean()
```

```python
from sklearn.linear_model import Lasso,Ridge,ElasticNet
```

```python
m_lasso = Lasso(alpha=0.8,max_iter=8000)
m_ridge = Ridge(alpha=0.8,max_iter=8000)
m_elastic = ElasticNet(alpha=0.8,l1_ratio=0.4,max_iter=8000)
m_lasso.fit(Xtrain_preprocess,ytrain)
m_ridge.fit(Xtrain_preprocess,ytrain)
m_elastic.fit(Xtrain_preprocess,ytrain)
```

```python
yl=m_lasso.predict(Xtrain_preprocess)
yr=m_ridge.predict(Xtrain_preprocess)
ye=m_elastic.predict(Xtrain_preprocess)
```

```python
print(f'The training score is {m.score(Xtrain_preprocess,ytrain)}')
print(f'R2 score for Linear regression {m.score(Xtrain_preprocess,ytrain)}')
print(f'R2 score for Ridge regression {m_ridge.score(Xtrain_preprocess,ytrain)}')
print(f'R2 score for Lasso regression {m_lasso.score(Xtrain_preprocess,ytrain)}')
print(f'R2 score for Elastic net regression {m_elastic.score(Xtrain_preprocess,ytrain)}')
```
